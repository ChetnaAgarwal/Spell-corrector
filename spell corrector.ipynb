```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
import string
punctuations = string.punctuation
punctuations
```

```python
p = [i for i in punctuations]
```

```python
words = []
def parse_line(l):
    b = word_tokenize(l)
    for i in b:
        i = i.lower()
        if i not in p:
            words.append(i)
```

```python
f1 = open("big.txt")
while True:
    line = f1.readline()
    parse_line(line)
    if ("" == line):
        break;
words
```

```python
# removing non alphabet characters from words
non_alpha = []
for i in range(0, len(words)):
    s = ''.join(ch for ch in words[i] if ch.isalpha())
    non_alpha.append(s)
non_alpha
non_alpha = [x for x in non_alpha if x] #removing empty strings
import pickle
with open('file1', 'wb') as fp:
    pickle.dump(non_alpha, fp)
non_alpha
```

```python
with open ('file1', 'rb') as fp:
    non_alpha = pickle.load(fp)
```

```python
# to find frequency of the word in the corpus, make a dictionary of words and their frequencies
count = pd.DataFrame(non_alpha, columns=['word'])
count['times'] = 1
count.head()
```

```python
count = count.groupby('word').aggregate('count')
count = pd.DataFrame(count['times'])
count = count.reset_index()
count.head()
```

```python
count = count.set_index('word')['times'].to_dict()
```

```python
def skip_one_letter_func(word):
    list = []
    #for i in substrings:
    for j in range(0, len(word)):
        list.append(word[:j] + word[j+1:])
    return list
```

```python
def adj_letters_int_func(word):
    list = []
    #for i in substrings:
    for j in range(0, len(word)-1):
        list.append(word[:j]+word[j+1]+word[j]+word[j+2:])
    return list
```

```python
def incorrect_letter_func(word):
    list = []
    chars = 'abcdefghijklmnopqrstuvwxyz'
    #for i in substrings:
    for j in range(0, len(word)):
        for ch in range(0, len(chars)):
            list.append(word[:j] + chars[ch] + word[j+1:])
    return list
```

```python
def extra_letter_func(word):
    list = []
    chars = 'abcdefghijklmnopqrstuvwxyz'
    #for i in substrings:
    for j in range(0, len(word)+1):
        for ch in range(0, len(chars)):
            list.append(word[:j]+chars[ch]+word[j:])
    return list
```

```python
# to find out incorrect variations of the word
def variations(word):
    #substrings = substrings_func(word)
    extra_letter = extra_letter_func(word) # words obtained by inserting one extra letter at random
    skip_one_letter = skip_one_letter_func(word) # words obtained by skipping one letter
    adj_letters_int = adj_letters_int_func(word) # words obtained by interchanging 2 adj letters 
    incorrect_letter = incorrect_letter_func(word) # words obtained by typing incorrect letter
    var = set(skip_one_letter + extra_letter+ adj_letters_int+ incorrect_letter)
    return [w for w in var if w in non_alpha]
```

```python
def max_prob_word(word):
    word_list = variations(word)
    #print(word_list)
    freq_list = []
    for i in range(0,len(word_list)):
        freq_list.append(count[word_list[i]])
    #print(freq_list)
    max_index = freq_list.index(max(freq_list))
    return word_list[max_index]
```

```python
def spellcorrect(word):
    if word not in non_alpha:
        try:
            return max_prob_word(word)
        except:
            print("couldnt find correct word")
    else:
        return word
```
